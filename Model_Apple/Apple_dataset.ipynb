```python
import numpy as np
import pandas as pd
import joblib
```

```python
df = pd.read_csv(r'C:\Users\batul\OneDrive\Desktop\ITR_Projects\Model_Apple\apple_products_pricing.csv')
```

```python
print(df.isnull().sum())
df.drop_duplicates(inplace=True)
```

```python
df = df.drop('Launch_Price_INR', axis=1)
```

```python
df.columns
```

```python
X = df.drop("Current_Price_INR",axis=1)
print(X)
Y = df["Current_Price_INR"]
```

```python
X = pd.get_dummies(X)
print(X)
```

```python
columns = X.columns
joblib.dump(columns, 'columns.pkl')
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
print(X_train.shape, X_test.shape, Y_train.shape, Y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
numerical_columns = ['Discount_Pct', 'Rating', 'Reviews_Count']
X_train[numerical_columns] = scaler.fit_transform(X_train[numerical_columns])
X_test[numerical_columns] = scaler.transform(X_test[numerical_columns])
print(X_train.head())
print(X_test.head())
joblib.dump(scaler, 'scaler.pkl')
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, Y_train)
joblib.dump(model, 'model.pkl')
```

```python
model.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2_score = r2_score(Y_test, model.predict(X_test))
print("R2 Score:", r2_score)
```
